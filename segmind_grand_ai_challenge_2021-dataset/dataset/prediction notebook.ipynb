```python
import pandas as pd
```

```python
train_data=pd.read_csv('TRAIN.csv')
test_data=pd.read_csv('TEST.csv')
store=pd.read_csv('store.csv')
```

```python
train_data.info()
```

```python
store.info()
```

```python
train_store=pd.merge(left=train_data,right=store,on=['Store'])
test_store=pd.merge(left=test_data,right=store,on=['Store'])
```

```python
train_store.head()
```

```python
## date rage for each store
pd.set_option("display.max_rows", None, "display.max_columns", None)
train_store_agg=train_store.groupby('Store')['Date'].agg(['min', 'max'])
train_store_agg
```

```python
test_store_agg=test_store.groupby('Store')['Date'].agg(['min', 'max'])
test_store_agg
```

```python
## basic EDA
def eda_table(df):
    sum_table=pd.DataFrame()
    for i in [i for i in df.columns]:
        attr_info=pd.DataFrame({
        'attribute':[i],
        'countofelements':[df[i].nunique()],
        'NA_Proportion':[df[i].isnull().sum()/len(df[i])]})
        sum_table=sum_table.append(attr_info)
    return sum_table 
eda_table(train_store)
```

```python
eda_table(test_store)
```
